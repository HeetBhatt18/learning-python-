#   **Methods Of Lists.**






```python
List = [10,20,10.5,"hello",True]
List
```

```python
List.append(30) # .append(elem) is used to add the values in the list. It will add the values from the last index.
List
```

```python
List.remove("hello") # .remove(elem) is used remove the values from the list. It will remove the values from the last index.
List
```

```python
List.insert(3,"Hello") # .insert(index, elem) is also used to add the values in the list just like append but, insert is used to insert the values at certain index.
List
```

```python
List.pop(1) # .pop(index) is used to remove values at certain index.
List
```

```python
List2 = ["World",True,445,33.9]
List2
```

```python
List.extend(List2) # .extend(list2) is used to merge two lists in one.
List
```

```python
List2.clear # .clear is used to remove all the elements from the list
List2
```

```python
Alphabets = ["f","i","m","a","v","n","b","c","g"] # .sort() is used arrange the elements in Ascending order. It can be Number, Alphabetical or string etc.
Alphabets.sort()
Alphabets
```

```python
Alphabets_rev = ["f","i","m","a","v","n","b","c","g"] # .reverse() is used to reverse the elements according to descending order.
Alphabets_rev.reverse()
Alphabets_rev
```

```python
Repeat = [10,20,30,[1,2,3],30,40,50] # .count(elem) is used to find out the values which are repeated in the list.
Repeat.count(30)
```

```python
New = [10, "my", 10.5 ,"name", True] # .copy() is used to return a shallow copy of the list.
New.copy()
```

# FOR and IN 

### For and In are the constructor used to look at each element in the list.

```python
Number = [10,20,30,40,50] 
sum = 0
for i in Number:
    sum = sum + i
sum 
```

# IF and IN 

### It is used to test if the values is in the collection of list. Returning, true or false.

```python
Message = ["hello","hi","welcome"]
if "hello" in Message:
    print("Yes, it is present in the list")
```
