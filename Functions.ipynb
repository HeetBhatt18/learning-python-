# **Functions**

### A function is a block of code that performs a specific task and runs only when called.

### Defined using def keyword

### Can return values using return 

### Functions improve code reusability



# **Types of functions**

# **1.Default argument function**

### Default arguments are evaluated once at function definition time

```python
def sum(a,b,c): 
    print("The sum is: ",a , b , c)

sum(10,20,30)

# Note :- non-default argument cannot be followed by default argument 
```

```python
def greet(name="Guest", message="Welcome!"):
    print(name + ", " + message)

greet()
greet("Heet")
greet("Heet, Good morning")
```

```python
def withdraw(amount, balance=1000):
    if amount <= balance:
        print("Withdraw Sucessfull")
    else:
        print("Insufficient balance")

withdraw(500)
withdraw(1500)
```

```python
# Smart Calculator

def calculator(a, b, operation="add"):
    if operation == "add":
        return a + b
    elif operation == "sub":
        return a - b
    elif operation == "mul":
        return a * b
    elif operation == "div":
        return a / b
    else:
        return "Invalid operation"

print(calculator(10, 5))          
print(calculator(10, 5, "mul"))   
```

# **2.Keyword argument function**

### 

```python
def student(name, age):
    print("Name:", name)
    print("Age:", age)

# Calling using keyword arguments

student(age=21,name="Heet")
```

```python
def calculate(price, discount):
    final_price = price - discount
    print("Final Price:", final_price)

calculate(discount=50, price=500)
```

```python
# Shopping Cart System 🛒 

def bill(item, price, quantity=1, discount=0):
    total = price * quantity
    final_amount = total - discount

    print("Item:", item)
    print("Total:", total)
    print("Discount:", discount)
    print("Final Amount:", final_amount)

# Calling the function

bill(item="Laptop", price=50000, discount=1500)
bill(item="Phone", price=20000, quantity=2)
bill(item="Headphones", price=3000, quantity=2, discount=500)
```

```python
# Student Result System 🎓

def result(name, marks1, marks2, marks3, grace=0):
    total = marks1 + marks2 + marks3 + grace
    percentage = total / 3

    print("Name:", name)
    print("Total:", total)
    print("Percentage:", percentage)

    if percentage >= 75:
        print("Grade: A")
    elif percentage >= 50:
        print("Grade: B")
    else:
        print("Grade: C")

# Calling the function

result(name="Heet", marks1=80, marks2=70, marks3=60)
result(name="Raj", marks1=40, marks2=45, marks3=50, grace=10)
```

# **3.Variable Function**

```python
def sum_x():
    sum = 0
    lists = [10,20,30,40]
    for i in range(4):
        sum += lists[i]

    return sum

print(sum_x())
```
