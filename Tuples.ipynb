#  **Tuples is a collection of Inhetrogenous data types.**

## Tuple is immutable, but we can use indexing and slicing in the tuples.

## Tuples is written in round brackets ().

## Tuples is a ordered collection.

# **Methods Of Tuples.**

```python
Tuple = (10,20,True,"hello",11.5,20)
Tuple
```

```python
Tuple.count(20) # .count(elem) is used to count the values which are repeated in the tuple.
```

```python
Tuple.count("hello")
```

```python
Tuple.count(True)
```

```python
Tuple_1 = ("hello","Python",True,10,10.5)
Tuple_1
```

```python
Tuple_1.index("Python") # .index(elem) is used to return that which element is present on which index.
```

```python
Tuple_1.index(10.5)
```

# Slicing

```python
Tup_1 = (10,20,30,"Hello",True,5.5)
Tup_1[0:]
```

```python
Tup_1[2:]
```

```python
Tup_1[0::2]
```
