# **Loop**

### Loop in python is used to repeat a block of code multiple times until a condition is satisfied.

### **Types of loops :**
#### 1. For loop - A for loop is used when you know how many times you want to repeat something.
#### 2. While loop - A while loop runs as long as the condition is True.
#### 3. Break - Stops the loop completely.
#### 4. Continue - Skips the current iteration.

# **For Loop**

```python
List = []
l = List
for i in range(6):
    a = int(input("Enter a number: "))
    l.append(a)
print(l)
```

```python
Words = input("Enter a string:")
for letters in Words:
    print(letters)
```

# **While Loop**

```python
i = 1

while i <= 5:
    print(i)
    i += 1
```

```python
i = 1
num = 12
while i <=10:
    print(num, "x", i, "=", num*i)
    i= i+1
```
