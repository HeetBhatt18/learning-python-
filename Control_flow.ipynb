# **Control Flow Tools**

### The control flow statements are if-else.

### The if statement can come seperately but else comes in pair with if or we can say that elif.

```python
num = int(input("Enter the number"))

if(num>0):
    print(f"The number is positive {num}")
elif(num<0):
    print(f"The number is negative {num}")
else:
    print(f"The number is zero {num}")
```

```python
# Nested if 

Marks = int(input("Enter the marks: "))

if(Marks<33):
    print("Fail")
if(Marks<= 75):
    print("The marks are good")
if(Marks>75):
    print("The marks are very good")
```

```python
pin = 1234
balance = 10000

entered_pin = int(input("Enter your PIN: "))

if entered_pin == pin:
    print("Login successful ✅")
    
    amount = int(input("Enter amount to withdraw: "))
    
    if amount <= balance:
        balance = balance - amount
        print("Transaction successful 💸")
        print(f"Remaining balance: {balance}")
    else:
        print("Insufficient balance ❌")

else:
    print("Incorrect PIN ❌")
```

```python
username = 'admin'
password = 1234

entered_user = input("Enter a username: ")

if entered_user == username:
    entered_pass = input("Enter a password: ")
    if entered_pass == password:
          print("Login successful ✅")
    else:
        print("Incorrect password ❌")

else:
    print("Invalid username ❌")
```
