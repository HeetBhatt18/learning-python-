# **Dictonary is a collection of hetrogenous elements.** 

## Dictonary is mutable, but we cannot use slicing and indexing.

## Dictonary are written within curly brackets { }. 

## Dictonary is a ordered collection.

## Dictonary syntax : {"Key1" : elements , "Key2" : elements}

# **Methods Of Dictonary.**

```python
Std_Id = {"name" : ["ajay","heet","vedant","harsh"], "age" : 21 , "course" : "deploma"} # The Key must be unique in the dictonary.
Std_Id.get("name") # .get is used to get the value of the key.
```

```python
Emp = {"Emp_id" : [11,12,13,14], "Emp_name" : ["Sumit","Amit","Satyam","Ajit"] , "Emp_salary" : [10000,20000,30000,40000]}
print(Emp.keys()) # .keys is used to get the keys of the dictonary.
print(Emp.values()) # .values is used to get the values of the dictonary
```

```python
Book = {"book_no" : 342 , "author_name" : "bala goswmi" , "book_name" : "C programming" , "price" : 1000}
Book
Book.update({"price" : 1200}) # .update is used to update the value of the key.
Book
```

```python
Book.pop("author_name") # .pop is used to remove the key and its value from the dictonary.
Book
```

```python
del Book ['book_no'] # .del is used to delete the key and its value from the dictonary.
Book
```

```python
len(Book) # .len is used to get the number of key-value pairs in the dictonary.
```

# **Nested Dictonary.**

### We can use tuple to make the key.

```python
data = {
    's1' : {'name': 'A','age':18}, 
    's2' : {'name': 'B','age':19},
}
data
```

```python
data['s1']['name'] # to access the value of the key in the nested dictonary.
```

```python
data.update({'s3' : {'name': 'C','age':20}}) # to add a new key-value pair in the nested dictonary.
data
```

```python
data['s1']['name']=['H','I'] # to update the value of the key in the nested dictonary.
data
```

```python
data['s2']['age']
```

```python
data.get('s1') # to get the value of the key in the nested dictonary.
```
