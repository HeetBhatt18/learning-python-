# **Methods of String**

```python
Name = "heet" # .upper() is a string method that converts all characters in the string to uppercase.
Name.upper()
```

```python
company = ("TATA MOTORS") # .lower() is a string method that converts all characters in the string to lowercase.
company.lower()
```

```python
first_name = "Heet"
last_name = "Bhatt"
full_name = first_name + " " + last_name # Concatenation of strings using the + operator.
full_name
```
