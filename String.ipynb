# **String is a collection of characters.**

### String is immutable, so slicing and indexing can be used.

### String is a ordered collection. 

# **Methods of String**

```python
Name = "heet" # .upper() is a string method that converts all characters in the string to uppercase.
Name.upper()
```

```python
company = ("TATA MOTORS") # .lower() is a string method that converts all characters in the string to lowercase.
company.lower()
```

```python
first_name = "Heet"
last_name = "Bhatt"
full_name = first_name + " " + last_name # Concatenation of strings using the + operator.
full_name
```

# **Formated string**

### An f-string is a way to insert variable and expression directly inside a string using {}.

```python
Name = "Heet"
Age = 21 

print(f"Hello, my name is {Name} and my age is {Age}.")
```

```python
a = 10
b = 20

print(f"The sum of {a} and {b} is: {a+b}")
```

```python
# Multiline f-string

name = "Heet"
age = 21

print(f"""
Name: {name}
Age: {age}
""")
```
